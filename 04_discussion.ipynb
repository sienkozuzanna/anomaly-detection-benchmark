## Discussion

**Isolation Forest** is the most practical algorithm overall - stable across dimensionalities, insensitive to hyperparameters, and competitive on both tabular and geometric data. It is the safest default when no labels are available.

**OC-SVM** has the highest ceiling but also the highest risk - kernel and gamma choice can make AUC-ROC differ by over 0.5 on a single dataset. With proper tuning it matches or beats everything else, without it, it is unreliable.

**LOF** is the right tool for low-dimensional data with local outlier structure, but the curse of dimensionality makes it unsuitable for high-feature tabular datasets. n_neighbors must always be swept - there is no safe default.

**DBSCAN** is the hardest to use: two strongly interacting parameters, density assumptions that rarely hold on real tabular data, and the highest failure count alongside LOF. It rewards careful tuning on geometric data but is a poor choice for general-purpose anomaly detection.

**ECOD** competitive on tabular data with zero tuning effort. Its weakness is local outlier structure.

**Overall** is that no single method dominates across all dataset types. Global methods (Isolation Forest, ECOD) generalise better across diverse data, local methods (LOF, DBSCAN) are more powerful in the right setting but require more effort and domain knowledge.
